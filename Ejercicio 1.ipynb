```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fiuba_4_postulaciones.csv')
df['fechapostulacion'] = pd.to_datetime(df['fechapostulacion'])
df['diasemana'] = df['fechapostulacion'].dt.weekday_name
```

```python
df.head()
```

```python
df['diasemana'] = df['fechapostulacion'].dt.weekday_name
```

```python
df.head()
```

```python
postulacionesdia = df.groupby('diasemana')['idaviso'].count().sort_values()
```

```python
postulacionesdia = df['diasemana'].value_counts()
```

```python
postulacionesdia
```

```python
postulacionesdia.plot(kind='bar',figsize=(16,8),rot=85,title='Postulaciones por dia');
ax=plt.gca()
ax.set_ylabel('Cantidad de postulaciones')
ax.set_xlabel('Dia de la semana')
```

Como podemos ver en el gráfico, los días donde hubo más postulaciones fueron los días Miércoles.

```python
df['mes'] = df['fechapostulacion'].dt.strftime('%b')
```

```python
df
```

```python
postulacionesmes = df.groupby('mes')['idaviso'].count().sort_values()
```

```python
postulacionesmes
```

```python
postulacionesmesdia = df.groupby(['mes', 'diasemana'])['idaviso'].count().sort_values()
```

```python
postulacionesmesdia
```
