```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fiuba_6_avisos_detalle.csv')
```

```python
df.head()
```

```python
areas = df.groupby('nombre_area')['idaviso'].count().sort_values()
```

```python
areas
```

```python
areasReduced = areas[areas>100]
```

```python
areasReduced
```

```python
areasReduced.plot(kind='bar',figsize=(16,8),rot=85,title='Postulaciones por Area');
ax=plt.gca()
ax.set_ylabel('Cantidad de postulaciones')
ax.set_xlabel('Dia de la semana')
```

El gráfico muestra claramente que el área de Ventas es donde se encuentra la mayor cantidad de ofertas laborales.

```python
zonaDemandaVentas = df['nombre_zona'][df['nombre_area'] == 'Ventas'].value_counts()
```

```python
zonaDemandaVentas
```

```python
zonaDemandaVentas.plot(kind='bar',figsize=(16,8),rot=85,title='Ofertas de Avisos para el área Ventas por Zona');
ax=plt.gca()
ax.set_ylabel('Cantidad de Ofertas Laborales')
ax.set_xlabel('Zona')
```

Del análisis realizado, vemos que las ofertas del área Ventas provienen mayoritariamente del Gran Buenos Aires.

```python
areasConZonas = df.groupby(['nombre_area', 'nombre_zona'])['idaviso'].count().sort_values()
```

```python
areasConZonasReduced = areasConZonas[areasConZonas > 50]
```

```python
areasConZonasReduced
```

```python
areasConZonasReduced.plot(kind='bar',figsize=(16,8),rot=85,title='Ofertas por Area y Zona');
ax=plt.gca()
ax.set_ylabel('Cantidad de Ofertas')
ax.set_xlabel('Area y Zona')
```

El ultimo gráfico muestra como se distribuyen las ofertas laborales de cada Área dentro de cada Zona.
